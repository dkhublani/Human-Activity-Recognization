```r
install.packages('class')
install.packages('readr')
```

```r
library(class)
library(readr)
X_train <- read_delim("/home/dupad/Desktop/613/project/appendix/X_train.txt", 
    "\t", escape_double = FALSE, trim_ws = TRUE)

y_train <- read_delim("/home/dupad/Desktop/613/project/appendix/y_train.txt", 
    "\t", escape_double = FALSE, trim_ws = TRUE)

X_test <- read_delim("/home/dupad/Desktop/613/project/appendix/X_test.txt", 
    "\t", escape_double = FALSE, trim_ws = TRUE)

y_test <- read_delim("/home/dupad/Desktop/613/project/appendix/y_test.txt", 
    "\t", escape_double = FALSE, trim_ws = TRUE)
train_data<-data.frame(Y1 = as.factor(y_train$Y1), X_train)
test_data<-data.frame(Y1 = as.factor(y_test$Y1),X_test)
```

```r
install.packages('caret')
```

```r
library(caret)
model<-pcaNNet(Y1~.,train_data,size=5,decay=5e-4, maxit=500)
```

```r
pred<-predict(model,test_data,type=c("class"))
mean(pred!=test_data$Y1)
```

```r
model<-pcaNNet(Y1~.,train_data,size=5,decay=5e-4, maxit=1000)
pred<-predict(model,test_data,type=c("class"))
mean(pred!=test_data$Y1)
```
